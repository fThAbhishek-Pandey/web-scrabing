```python
from bs4 import BeautifulSoup
```

read html file

```python
with open("index.html",'r') as html_file:
  content=html_file.read()
  print(content)
```

```python
soup=BeautifulSoup(content,'lxml')
print(soup.prettify())
```

```python
head= soup.find("head") # return first tag
print(head)
```

```python
div_tag = soup.find_all("div")
print(div_tag)
```

```python
for i in div_tag:
  print(i)
```

```python
for i in div_tag:
  print(i.text)
```

```python
slok = soup.find_all('div', 'div',class_="slok-text")
print(slok)
```

```python
#print(slok.text)
for i in slok:
  print(i.text)
```

```python
benifits= soup.find_all('div','div','div',class_="beni-title")
print(benifits)
```

```python
for i in benifits:
  print(i.text)
```

```python
beni_num= soup.find_all('div','div','div',class_="beni-num")
#print(beni_num.text)
for i in beni_num:
  print(i.text)
```
